# 4. Data preprocessing

```python
df.shape
```

```python
# Split data into training and testing datasets
data_train = pd.DataFrame(df.Close.AAPL[0 : int(len(df) * 0.7)])
data_test = pd.DataFrame(df.Close.AAPL[int(len(df) * 0.7) : int(len(df))])
data_train, data_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
data_train_scale = scaler.fit_transform(data_train)
```

```python
data_train_scale
```

# 5. Sequence Creation

```python
def sequence_creation(data):
    x_train = []
    y_train = []
    for i in range(100, data.shape[0]):
        x_train.append(data[i - 100 : i])
        y_train.append(data[i, 0])
    return np.array(x_train), np.array(y_train)
```

```python
x_train, y_train = sequence_creation(data_train_scale)
```

```python
x_train
```

```python
y_train
```

# 6. Model Building

```python
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Input
```

```python
model = Sequential()
model.add(Input(shape=(100, 1)))
model.add(LSTM(units=128, activation='tanh', return_sequences=True))
model.add(LSTM(units=64))
model.add(Dense(25))
model.add(Dense(1))

model
```

# 7. Model Training

```python
model.compile(optimizer='adam', loss='mse')
model.fit(x_train, y_train, epochs=50)
```

```python
model.summary()
```

```python
# Save trained model
model.save("stock_prediction_model.keras")
```

# 8. Prepare testing data

```python
data_train
```

```python
past = data_train.tail(100)
past
```

```python
dat_test = pd.concat([past, data_test], ignore_index=True)
dat_test
```

```python
data_test_scale = scaler.fit_transform(dat_test)
data_test_scale
```

```python
x_test, y_test = sequence_creation(data_test_scale)
x_test, y_test
```

```python
x_test[0].shape
```

# 9. Make prediction

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
y_predicted = scaler.inverse_transform(y_predicted.reshape(-1, 1)).flatten()
y_test = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
```

```python
y_predicted.shape
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label="Original price")
plt.plot(y_predicted, label="Predicted price")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
```

# 10. Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_predicted)
mse
```
